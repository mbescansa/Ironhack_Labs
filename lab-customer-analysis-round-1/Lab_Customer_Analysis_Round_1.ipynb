#### 1. Open files and show shape of each one

```python
# import data from links
url1 = 'https://raw.githubusercontent.com/ironhack-labs/lab-customer-analysis-round-1/master/files_for_lab/csv_files/file1.csv'
url2 = 'https://raw.githubusercontent.com/ironhack-labs/lab-customer-analysis-round-1/master/files_for_lab/csv_files/file2.csv'
url3 = 'https://raw.githubusercontent.com/ironhack-labs/lab-customer-analysis-round-1/master/files_for_lab/csv_files/file3.csv'

table1=pd.read_csv(url1)
table2=pd.read_csv(url2)
table3=pd.read_csv(url3)

table1.shape
table2.shape
table3.shape

table1.head()
table2.head()
table3.head()
```

#### 2. Standarize header names

We rename the columns that are wrong.

```python
# We rename ST and GENDER columns on table1 and table2
table1.rename(columns={"ST": "State", 'GENDER':'Gender'}, inplace=True)
table2.rename(columns={"ST": "State", 'GENDER':'Gender'}, inplace=True)
table1.head(1)
table2.head(1)
table3.head(1)
```

Now we specify the exact order in which we want the columns, following the order of table 1.

```python
table1.columns
```

```python
# We get all column names for all tables with the same format and order 
table2=table2[['Customer', 'State', 'Gender', 'Education', 'Customer Lifetime Value',
       'Income', 'Monthly Premium Auto', 'Number of Open Complaints',
       'Policy Type', 'Vehicle Class', 'Total Claim Amount']]
table3=table3[['Customer', 'State', 'Gender', 'Education', 'Customer Lifetime Value',
       'Income', 'Monthly Premium Auto', 'Number of Open Complaints',
       'Policy Type', 'Vehicle Class', 'Total Claim Amount']]
table1.head(1)
table2.head(1)
table3.head(1)
```

#### 3. Concatenate all tables

```python
# Customer Lifetime Value ends with a % sign on table1 and table2. We will erase it with a function.
def remove_percentage_sign(x):
  if type(x)==str:
    return x.replace("%"," ")

  else:
    return float(x)
```

```python
# Erase percentage on both tables. Now all three tables have the same format for Customer Lifetime Value
table1['Customer Lifetime Value'] = list(map(remove_percentage_sign,table1['Customer Lifetime Value']))
table1['Customer Lifetime Value'].head()
table2['Customer Lifetime Value'] = list(map(remove_percentage_sign,table2['Customer Lifetime Value']))
table2['Customer Lifetime Value'].head()
```

```python
# Customer Lifetime Value is an object type on table 1 and table 2, so we will convert them to float now that the % sign has been removed
def conv_float(x):
  if type(x) == str:
    return float(x)
  else:
    return 
```

```python
# Now all tables have the same data type and format for the column Customer Lifetime Value
table1['Customer Lifetime Value'] = list(map(conv_float,table1['Customer Lifetime Value']))
table1['Customer Lifetime Value'].head()
table2['Customer Lifetime Value'] = list(map(conv_float,table2['Customer Lifetime Value']))
table2['Customer Lifetime Value'].head()
```

```python
table3['Customer Lifetime Value'].head()
```

```python
mergedData=pd.concat([table1,table2,table3], axis=0)
mergedData.shape
```
